```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_count = list(df['brand'].value_counts().values)
no_of_count
```

```python
plt.figure(figsize=(10,4))              # (widh , height)
brand_name = list(df['brand'].value_counts().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,ec ='black',color = 'red')
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.ylabel('no. of bikes   --->')
plt.xlabel('brand name   --->')
plt.savefig('brandpopulation')
plt.show()
```
